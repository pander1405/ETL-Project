```python
import pandas as pd
```

```python
df = pd.read_csv("bestsellers_with_categories.csv")

df.head()
```

```python
df_fiction = df.loc[(df['Genre'] == "Fiction") & (df['Year']==2019),]

df_fiction = df_fiction.drop(columns="Price")
```

```python
df_fiction.shape
```

```python
df_fiction = df_fiction.reset_index(drop=True)
```

```python
df_fiction.to_csv("2019_bestsellers_from_csv.csv")
```

```python
df_fiction
```

```python
authors = df_fiction["Author"]
authors = pd.DataFrame(authors)

first_names = []
last_names = []

for a in authors["Author"]:
    name = a.split(" ")
    if len(name) > 2:
        if "Jr." in name:
            first = name[0]
            last = name[1]+ " "+ name[2]
        elif "." in name[1]:
            first = name[0] + " " + name[1]
            last = name[2]
        elif "Dr." in name[0]:
            first = name[0] + " " + name[1]
            last = ""
        else:
            first = name[0]
            last = name[1]+ " "+ name[2]
    else:
        first = name[0]
        last = name[1]
    
    first_names.append(first)
    last_names.append(last)


author_names = pd.DataFrame({"First":first_names, "Last":last_names})
author_names.to_csv("2019_Amazon-Best-Sellers_AuthorNamesOnly.csv")
author_names
```

```python
author_names["Book_Title"] = df_fiction["Name"]
author_names

author_names.to_csv("2019_Amazon-Best-Sellers_Author-BookTitles.csv")

author_names
```

```python
#full table with number of reviews and rating, author name split into first and last

author_names["UserRating"] = df_fiction["User Rating"]
author_names["Num_Reviews"] = df_fiction["Reviews"]
author_names["Year"] = df_fiction["Year"]
author_names["Genre"] = df_fiction["Genre"]

author_names.to_csv("2019_Amazon-Best-Sellers_FullTable_AuthorNamesSplit.csv")

author_names
```
